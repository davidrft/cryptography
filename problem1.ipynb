```python
def strxor(a, b):     # xor two strings of different lengths
    if len(a) > len(b):
       return "".join([chr(ord(x) ^ ord(y)) for (x, y) in zip(a[:len(b)], b)])
    else:
       return "".join([chr(ord(x) ^ ord(y)) for (x, y) in zip(a, b[:len(a)])])

key = [None]*150
knownKeyPositions = set()

for i, cypher1 in enumerate(cyphers):
    counter = collections.Counter()
    
    for j, cypher2 in enumerate(cyphers):
        if i == j: continue
        text = strxor(cypher1, cypher2)
        for index, char in enumerate(text):
            if char.isalpha(): counter[index] += 1
                
    knownSpaceIndexes = []
    
    for j, value in counter.items():
        if value >= 5: knownSpaceIndexes.append(j)
    
    xoredSpace = strxor(cypher1, ' '*150)
    
    for j in knownSpaceIndexes:
        key[j] = xoredSpace[j]
        knownKeyPositions.add(j)
        
key_ = ''.join([val if val is not None else '\0' for val in key])

output = strxor(key_, target)

print(''.join([char if index in knownKeyPositions else '*' for index, char in enumerate(output)])+"\n")
```
